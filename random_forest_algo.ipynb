```python
import pandas as pd
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
    plt.title(f"label: {digits.target[i]}")
    plt.show()
```
